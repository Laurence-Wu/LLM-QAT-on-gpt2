# Matplotlib Complete Tutorial
## A Comprehensive Guide to Data Visualization in Python

This notebook covers the essential grammar and functions of matplotlib for creating professional visualizations.

## 1. Introduction and Setup

Matplotlib is Python's most popular plotting library. It provides both a MATLAB-like interface (pyplot) and an object-oriented interface.

```python
# Essential imports
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# For inline plots in Jupyter
%matplotlib inline

# For interactive plots (optional)
# %matplotlib widget

# Set default figure size
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['figure.dpi'] = 80
```

## 2. Basic Plotting - The pyplot Interface

### 2.1 Simple Line Plot

```python
# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Simple plot
plt.plot(x, y)
plt.title('Simple Sine Wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

### 2.2 Multiple Lines with Styling

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

# Plot with different styles
plt.plot(x, y1, 'b-', label='sin(x)', linewidth=2)      # blue solid line
plt.plot(x, y2, 'r--', label='cos(x)', linewidth=2)     # red dashed line
plt.plot(x, y3, 'g-.', label='sin(x)cos(x)', alpha=0.7) # green dash-dot, transparent

plt.title('Trigonometric Functions')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)
plt.show()
```

## 3. Object-Oriented Interface (Recommended for Complex Plots)

### 3.1 Figure and Axes Objects

```python
# Create figure and axes explicitly
fig, ax = plt.subplots(figsize=(10, 6))

# Plot on the axes
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# Customize using axes methods
ax.set_title('Using Object-Oriented Interface', fontsize=16, fontweight='bold')
ax.set_xlabel('X axis', fontsize=12)
ax.set_ylabel('Y axis', fontsize=12)
ax.legend()
ax.grid(True, linestyle=':', alpha=0.6)

# Set axis limits
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

plt.show()
```

### 3.2 Multiple Subplots

```python
# Create 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Generate data
x = np.linspace(0, 2*np.pi, 100)

# Plot 1: Line plot
axes[0, 0].plot(x, np.sin(x), 'b-')
axes[0, 0].set_title('Line Plot')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('sin(x)')

# Plot 2: Scatter plot
axes[0, 1].scatter(x[::5], np.cos(x[::5]), c=x[::5], cmap='viridis')
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('cos(x)')

# Plot 3: Bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, 5)
axes[1, 0].bar(categories, values, color=['red', 'green', 'blue', 'yellow', 'purple'])
axes[1, 0].set_title('Bar Plot')
axes[1, 0].set_ylabel('Values')

# Plot 4: Histogram
data = np.random.normal(100, 20, 1000)
axes[1, 1].hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axes[1, 1].set_title('Histogram')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')

# Adjust spacing between subplots
plt.tight_layout()
plt.show()
```

## 4. Different Plot Types

### 4.1 Scatter Plots

```python
# Generate random data
np.random.seed(42)
n = 150
x = np.random.randn(n)
y = 2 * x + np.random.randn(n) * 0.5
colors = np.random.rand(n)
sizes = np.random.uniform(20, 200, n)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Basic scatter
ax1.scatter(x, y, alpha=0.5)
ax1.set_title('Basic Scatter Plot')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Advanced scatter with colors and sizes
scatter = ax2.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
ax2.set_title('Scatter with Colors and Sizes')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
plt.colorbar(scatter, ax=ax2, label='Color Scale')

plt.tight_layout()
plt.show()
```

### 4.2 Bar Plots and Histograms

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Vertical bar plot
categories = ['Product A', 'Product B', 'Product C', 'Product D']
values = [23, 45, 56, 78]
axes[0, 0].bar(categories, values, color='steelblue')
axes[0, 0].set_title('Vertical Bar Plot')
axes[0, 0].set_ylabel('Sales')

# Horizontal bar plot
axes[0, 1].barh(categories, values, color='coral')
axes[0, 1].set_title('Horizontal Bar Plot')
axes[0, 1].set_xlabel('Sales')

# Grouped bar plot
x_pos = np.arange(len(categories))
values1 = [23, 45, 56, 78]
values2 = [35, 38, 62, 65]
width = 0.35

axes[1, 0].bar(x_pos - width/2, values1, width, label='2023', color='lightblue')
axes[1, 0].bar(x_pos + width/2, values2, width, label='2024', color='lightcoral')
axes[1, 0].set_xticks(x_pos)
axes[1, 0].set_xticklabels(categories)
axes[1, 0].set_title('Grouped Bar Plot')
axes[1, 0].legend()

# Stacked bar plot
axes[1, 1].bar(categories, values1, label='Q1', color='skyblue')
axes[1, 1].bar(categories, values2, bottom=values1, label='Q2', color='orange')
axes[1, 1].set_title('Stacked Bar Plot')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

### 4.3 Pie Charts and Donuts

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Data
labels = ['Python', 'JavaScript', 'Java', 'C++', 'Other']
sizes = [35, 25, 20, 15, 5]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.1, 0, 0, 0, 0)  # explode first slice

# Pie chart
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.set_title('Programming Languages Usage')

# Donut chart
wedges, texts, autotexts = ax2.pie(sizes, labels=labels, colors=colors, 
                                     autopct='%1.1f%%', startangle=90,
                                     pctdistance=0.85)

# Create circle for donut hole
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
ax2.add_artist(centre_circle)
ax2.set_title('Donut Chart')

plt.tight_layout()
plt.show()
```

### 4.4 Box Plots and Violin Plots

```python
# Generate sample data
np.random.seed(42)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(110, 15, 200)
data4 = np.random.normal(95, 25, 200)
all_data = [data1, data2, data3, data4]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Box plot
bp = ax1.boxplot(all_data, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'],
                  patch_artist=True)

# Color the boxes
colors = ['lightblue', 'lightgreen', 'lightcoral', 'lightyellow']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

ax1.set_title('Box Plot')
ax1.set_ylabel('Values')
ax1.grid(True, alpha=0.3)

# Violin plot
parts = ax2.violinplot(all_data, positions=[1, 2, 3, 4], 
                        showmeans=True, showmedians=True)

for pc, color in zip(parts['bodies'], colors):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

ax2.set_title('Violin Plot')
ax2.set_ylabel('Values')
ax2.set_xticks([1, 2, 3, 4])
ax2.set_xticklabels(['Group 1', 'Group 2', 'Group 3', 'Group 4'])
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Advanced Customization

### 5.1 Colors, Markers, and Line Styles

```python
x = np.linspace(0, 10, 20)

fig, ax = plt.subplots(figsize=(12, 8))

# Different line styles
line_styles = ['-', '--', '-.', ':']
markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*', 'h']
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange', 'purple', 'brown']

for i in range(10):
    y = np.sin(x + i * 0.5) * (i + 1) / 5
    ax.plot(x, y, 
            color=colors[i], 
            linestyle=line_styles[i % 4],
            marker=markers[i],
            markersize=8,
            markerfacecolor='white',
            markeredgecolor=colors[i],
            markeredgewidth=2,
            linewidth=2,
            label=f'Style {i+1}',
            alpha=0.8)

ax.set_title('Line Styles, Markers, and Colors', fontsize=16, fontweight='bold')
ax.set_xlabel('X axis', fontsize=12)
ax.set_ylabel('Y axis', fontsize=12)
ax.legend(loc='upper left', bbox_to_anchor=(1, 1), ncol=2)
ax.grid(True, alpha=0.3, linestyle=':')

plt.tight_layout()
plt.show()
```

### 5.2 Annotations and Text

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, y, 'b-', linewidth=2, label='sin(x)')

# Add annotations
ax.annotate('Maximum', 
            xy=(np.pi/2, 1), 
            xytext=(np.pi/2 + 1, 1.3),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=12, 
            color='red',
            fontweight='bold')

ax.annotate('Minimum', 
            xy=(3*np.pi/2, -1), 
            xytext=(3*np.pi/2 + 1, -1.3),
            arrowprops=dict(arrowstyle='->', color='green', lw=2),
            fontsize=12, 
            color='green',
            fontweight='bold')

# Add text
ax.text(np.pi, 0, '', fontsize=16, ha='center', va='bottom')
ax.text(2*np.pi, 0, '2', fontsize=16, ha='center', va='bottom')

# Add horizontal and vertical lines
ax.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
ax.axvline(x=np.pi, color='gray', linestyle='--', alpha=0.5)
ax.axvline(x=2*np.pi, color='gray', linestyle='--', alpha=0.5)

# Shade an area
ax.fill_between(x, 0, y, where=(x >= np.pi/4) & (x <= 3*np.pi/4), 
                 alpha=0.3, color='yellow', label='Shaded area')

ax.set_title('Annotations and Text Example', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### 5.3 Colormaps and Contour Plots

```python
# Create grid data
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Contour plot
contour = axes[0, 0].contour(X, Y, Z, levels=15, cmap='viridis')
axes[0, 0].clabel(contour, inline=True, fontsize=8)
axes[0, 0].set_title('Contour Plot')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')

# Filled contour
contourf = axes[0, 1].contourf(X, Y, Z, levels=20, cmap='plasma')
fig.colorbar(contourf, ax=axes[0, 1], label='Z value')
axes[0, 1].set_title('Filled Contour Plot')
axes[0, 1].set_xlabel('X')
axes[0, 1].set_ylabel('Y')

# Heatmap
im = axes[1, 0].imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='coolwarm', aspect='auto')
fig.colorbar(im, ax=axes[1, 0], label='Z value')
axes[1, 0].set_title('Heatmap (imshow)')
axes[1, 0].set_xlabel('X')
axes[1, 0].set_ylabel('Y')

# 3D-like surface plot (using pcolormesh)
pcm = axes[1, 1].pcolormesh(X, Y, Z, shading='gouraud', cmap='RdBu')
fig.colorbar(pcm, ax=axes[1, 1], label='Z value')
axes[1, 1].set_title('Pcolormesh Plot')
axes[1, 1].set_xlabel('X')
axes[1, 1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```

## 6. Working with Real Data - Loss Curves Example

### 6.1 Loading and Plotting Training Losses

```python
# Simulate training loss data (replace with your actual data)
np.random.seed(42)
epochs = np.arange(1, 101)
train_loss = 10 * np.exp(-epochs/20) + np.random.normal(0, 0.1, 100) + 2
val_loss = 10 * np.exp(-epochs/25) + np.random.normal(0, 0.15, 100) + 2.5

# Create figure with custom style
fig, ax = plt.subplots(figsize=(12, 6))

# Plot with smoothing
ax.plot(epochs, train_loss, 'b-', alpha=0.3, label='Train Loss (raw)')
ax.plot(epochs, val_loss, 'r-', alpha=0.3, label='Val Loss (raw)')

# Add smoothed lines (moving average)
window = 5
train_smooth = np.convolve(train_loss, np.ones(window)/window, mode='valid')
val_smooth = np.convolve(val_loss, np.ones(window)/window, mode='valid')
epochs_smooth = epochs[window-1:]

ax.plot(epochs_smooth, train_smooth, 'b-', linewidth=2, label='Train Loss (smoothed)')
ax.plot(epochs_smooth, val_smooth, 'r-', linewidth=2, label='Val Loss (smoothed)')

# Find and mark best validation loss
best_epoch = np.argmin(val_loss)
best_val_loss = val_loss[best_epoch]

ax.scatter(epochs[best_epoch], best_val_loss, color='green', s=100, zorder=5)
ax.annotate(f'Best Val Loss: {best_val_loss:.3f}\nEpoch: {epochs[best_epoch]}',
            xy=(epochs[best_epoch], best_val_loss),
            xytext=(epochs[best_epoch]+10, best_val_loss+0.5),
            arrowprops=dict(arrowstyle='->', color='green'),
            fontsize=10,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

# Styling
ax.set_xlabel('Epoch', fontsize=12)
ax.set_ylabel('Loss', fontsize=12)
ax.set_title('Training and Validation Loss Over Time', fontsize=14, fontweight='bold')
ax.legend(loc='upper right')
ax.grid(True, alpha=0.3, linestyle=':')
ax.set_xlim(0, 100)
ax.set_ylim(bottom=0)

# Add minor gridlines
ax.minorticks_on()
ax.grid(which='minor', alpha=0.1, linestyle=':')

plt.tight_layout()
plt.show()
```

## 8. Style Sheets and Customization

### 8.1 Using Built-in Styles

```python
# List available styles
print("Available styles:")
print(plt.style.available[:10])  # Show first 10 styles

# Create sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Try different styles
styles = ['default', 'seaborn', 'ggplot', 'fivethirtyeight']

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, style in enumerate(styles):
    with plt.style.context(style):
        axes[idx].plot(x, y1, label='sin(x)')
        axes[idx].plot(x, y2, label='cos(x)')
        axes[idx].set_title(f'Style: {style}')
        axes[idx].legend()
        axes[idx].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Set custom parameters
plt.rcParams.update({
    'figure.figsize': (10, 6),
    'font.size': 12,
    'font.family': 'serif',
    'axes.labelsize': 14,
    'axes.titlesize': 16,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'legend.fontsize': 11,
    'lines.linewidth': 2,
    'lines.markersize': 8,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'grid.linestyle': ':'
})

# Create plot with custom settings
fig, ax = plt.subplots()
x = np.linspace(0, 10, 50)
ax.plot(x, np.sin(x), 'o-', label='sin(x)')
ax.plot(x, np.cos(x), 's-', label='cos(x)')
ax.set_title('Plot with Custom RC Parameters')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.legend()
plt.show()

# Reset to defaults
plt.rcParams.update(plt.rcParamsDefault)
```

## 9. Interactive Features

### 9.1 Twin Axes

```python
# Data with different scales
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * 100
y2 = np.exp(x/2)

fig, ax1 = plt.subplots(figsize=(10, 6))

# First y-axis
color = 'tab:blue'
ax1.set_xlabel('X data')
ax1.set_ylabel('Sin values', color=color)
line1 = ax1.plot(x, y1, color=color, label='sin(x) * 100')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, alpha=0.3)

# Second y-axis
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Exp values', color=color)
line2 = ax2.plot(x, y2, color=color, label='exp(x/2)')
ax2.tick_params(axis='y', labelcolor=color)

# Combined legend
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left')

ax1.set_title('Plot with Twin Axes', fontsize=14, fontweight='bold')
fig.tight_layout()
plt.show()
```

## 10. Common Patterns and Best Practices

### 10.1 Complete Example: Publication-Ready Figure

```python
# Set publication style
plt.style.use('seaborn-v0_8-paper')

# Generate data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y_true = 2 * x + 1
y_measured = y_true + np.random.normal(0, 2, 100)
y_model = 2.1 * x + 0.8

# Create figure
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), 
                                gridspec_kw={'height_ratios': [3, 1]})

# Main plot
ax1.scatter(x, y_measured, alpha=0.5, s=20, label='Measured Data', color='gray')
ax1.plot(x, y_true, 'k-', linewidth=2, label='True Relationship')
ax1.plot(x, y_model, 'r--', linewidth=2, label='Model Fit')

# Add confidence interval
ci = 1.96 * 2  # 95% confidence interval
ax1.fill_between(x, y_model - ci, y_model + ci, alpha=0.2, color='red', 
                  label='95% Confidence Interval')

ax1.set_ylabel('Y Value', fontsize=12)
ax1.set_title('Model Fitting with Residuals Analysis', fontsize=14, fontweight='bold')
ax1.legend(loc='upper left', frameon=True, shadow=True)
ax1.grid(True, alpha=0.3, linestyle=':')

# Residual plot
residuals = y_measured - y_model
ax2.scatter(x, residuals, alpha=0.5, s=20, color='gray')
ax2.axhline(y=0, color='k', linestyle='-', linewidth=1)
ax2.axhline(y=np.mean(residuals), color='r', linestyle='--', linewidth=1, 
            label=f'Mean: {np.mean(residuals):.2f}')

# Add standard deviation bands
std_res = np.std(residuals)
ax2.axhline(y=std_res, color='b', linestyle=':', alpha=0.5)
ax2.axhline(y=-std_res, color='b', linestyle=':', alpha=0.5, 
            label=f'1 STD: {std_res:.2f}')

ax2.set_xlabel('X Value', fontsize=12)
ax2.set_ylabel('Residuals', fontsize=12)
ax2.legend(loc='upper right', fontsize=10)
ax2.grid(True, alpha=0.3, linestyle=':')

# Add text box with statistics
textstr = f'R = 0.95\nRMSE = {np.sqrt(np.mean(residuals**2)):.2f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.8)
ax1.text(0.02, 0.98, textstr, transform=ax1.transAxes, fontsize=10,
         verticalalignment='top', bbox=props)

plt.tight_layout()
plt.show()
```

## Summary of Key Functions

### Basic Plotting
- `plt.plot()` - Line plots
- `plt.scatter()` - Scatter plots
- `plt.bar()` / `plt.barh()` - Bar plots
- `plt.hist()` - Histograms
- `plt.pie()` - Pie charts
- `plt.boxplot()` - Box plots
- `plt.violinplot()` - Violin plots

### Figure and Axes
- `plt.figure()` - Create figure
- `plt.subplots()` - Create figure with subplots
- `ax.set_xlabel()` / `ax.set_ylabel()` - Set axis labels
- `ax.set_title()` - Set title
- `ax.legend()` - Add legend
- `ax.grid()` - Add grid
- `ax.set_xlim()` / `ax.set_ylim()` - Set axis limits

### Customization
- `plt.style.use()` - Apply style
- `plt.rcParams` - Modify default parameters
- `ax.annotate()` - Add annotations
- `ax.text()` - Add text
- `ax.axhline()` / `ax.axvline()` - Add horizontal/vertical lines
- `ax.fill_between()` - Fill area between curves

### Saving
- `plt.savefig()` - Save figure to file
- `plt.show()` - Display figure
- `plt.close()` - Close figure

### Color and Style Parameters
- Colors: 'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w' or hex codes
- Line styles: '-', '--', '-.', ':'
- Markers: 'o', 's', '^', 'v', 'D', 'p', '*', 'h', '+', 'x'
- Colormaps: 'viridis', 'plasma', 'coolwarm', 'RdBu', 'jet', etc.
